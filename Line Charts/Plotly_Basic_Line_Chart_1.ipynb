```python
import pandas as pd 
import numpy as np
%matplotlib inline
import plotly.graph_objects as go
import plotly.express as px
```

```python
arr_1=np.random.randn(50,4)
df_1=pd.DataFrame(arr_1,columns=['A','B','C','D'])
df_1.head()
```

Basic Matplotlib plot

```python
df_1.plot(figsize=(12,7));
```

Plotly Basic Line Chart

```python
px.line(df_1,x=list(df_1.index),y=['A','B','C','D'],labels={'x':'Index','value':'Data'})
```
